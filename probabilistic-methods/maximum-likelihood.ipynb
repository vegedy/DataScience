## $\hat{\mu}^{\text{MLS}}$ vs. $\bar{X}_N$

| Begriff                 | Bedeutung                                                    | Zufallsvariable? |
|------------------------|--------------------------------------------------------------|------------------|
| $\hat{\mu}^{\text{MLS}}$ | Konkreter Schätzwert aus beobachteten Daten $x_1, \dots, x_N$ | ❌ Nein           |
| $\bar{X}_N$              | Stichprobenmittel als Zufallsvariable basierend auf $X_1, \dots, X_N$ | ✅ Ja             |

### Erläuterung:

- $\hat{\mu}^{\text{MLS}} = \frac{1}{N} \sum_{i=1}^N x_i$:  
  Wird **nach Beobachtung** einer konkreten Stichprobe berechnet. Ergebnis ist eine **feste Zahl**.

- $\bar{X}_N = \frac{1}{N} \sum_{i=1}^N X_i$:  
  Ist eine **Zufallsvariable**, da sie von den Zufallsvariablen $X_i$ abhängt. Dient der **theoretischen Analyse** (z. B. Erwartungstreue, Varianz).

### Anwendung:

- Für **konkrete Berechnungen** mit Daten → $\hat{\mu}^{\text{MLS}}$
- Für **theoretische Eigenschaften** des Schätzers → $\bar{X}_N$


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Simulation: Stichprobe aus einer Normalverteilung
mu_true = 5
sigma_true = 2
N = 1000

# Zufallsvariablen X_i ~ N(mu_true, sigma_true)
X = np.random.normal(mu_true, sigma_true, N)

# Berechneter Schätzer aus realisierten Werten (nicht-zufällig)
mu_hat_mls = np.mean(X)

# Wiederholung zur Visualisierung der Verteilung von \bar{X}_N
replications = 5000
means = [np.mean(np.random.normal(mu_true, sigma_true, N)) for _ in range(replications)]

# Daten für Visualisierung
df = pd.DataFrame({
    "Wiederholte Mittelwerte (Zufallsvariable)": means
})

# Plot
plt.figure(figsize=(10, 5))
sns.histplot(means, kde=True, bins=50)
plt.axvline(mu_hat_mls, color='red', linestyle='--', label=f'Aktuelles $\hat{{\mu}}^{{MLS}}$ = {mu_hat_mls:.2f}')
plt.axvline(mu_true, color='green', linestyle=':', label=f'True $\mu$ = {mu_true}')
plt.title("Verteilung der Zufallsvariablen $\overline{X}_N$ bei mehrfacher Stichprobenziehung")
plt.xlabel("Stichprobenmittel")
plt.ylabel("Häufigkeit")
plt.legend()
plt.tight_layout()
plt.show()
```
