# Profilplot

Je weiter du auf der x-Achse gehst: Du gehst durch deine Klassen, also z. B. von kleinen zu großen Menschen.

Die Linie zeigt den typischen y-Wert in dieser x-Klasse (z. B. das durchschnittliche Gewicht).

Die Fehlerbalken zeigen, wie stark das Gewicht innerhalb der Klasse schwankt.

➡️ Es ist wie ein verdichtetes Streudiagramm: keine Punkte, sondern zusammengefasste Information pro Klasse.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binned_statistic

# Beispieldaten
np.random.seed(0)
x = np.random.uniform(150, 190, 1000) # Körpergröße
y = 0.4 * x + np.random.normal(0, 10, size=x.shape) # Gewicht

# Klassen definieren
bins = np.linspace(150, 190, 9) # 8 Klassen
bin_centers = 0.5 * (bins[1:] + bins[:-1])

mean_y, _, _ = binned_statistic(x, y, statistic='mean', bins=bins)
std_y, _, _ = binned_statistic(x, y, statistic='std', bins=bins)
```

```python
# Plot
plt.errorbar(bin_centers, mean_y, yerr=std_y, fmt='o', capsize=5, label='Mittelwert ± Std. Abweichung')
plt.xlabel('Körpergröße (cm)')
plt.ylabel('Gewicht (kg)')
plt.title('Profilplot mit Mittelwert und Standardabweichung')
plt.grid(True)
plt.legend()
plt.show()
```
