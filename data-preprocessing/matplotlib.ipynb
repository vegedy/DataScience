# Matplotlib

## Import libraries

```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.stats as stats
```

## Basic plot

```python
mu = 0
std = 1
x = np.linspace(start=-4, stop=4, num=100)
y = stats.norm.pdf(x, mu, std)
```

```python
plt.plot(x, y)
plt.show()
```

## Grid lines

```python
plt.plot(x, y)
plt.grid(color='red')
```

```python
plt.plot(x, y)

# major grid lines
plt.grid(which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=1.5)
# minor grid lines
plt.minorticks_on()
plt.grid(which='minor', color='beige', alpha=0.8, ls='-', lw=1)
```

## Labels & Axis

```python
plt.plot(x, y)
plt.xlabel('x', fontsize=15)
plt.ylabel('PDF', fontsize=15)
```

```python
plt.plot(x, y)
plt.xlim(-2, 2)
plt.ylim(0.1, 0.5)
```

```python
plt.plot(x, y)
plt.tick_params(axis='both', direction='out', length=6, width=2, labelcolor='b', colors='r', grid_color='gray', grid_alpha=0.5)
```

## Legend

```python
x = np.linspace(0, 7, 100)
line1, = plt.plot(x, np.sin(x), label='sin')
line2, = plt.plot(x, np.cos(x), label='cos')

plt.legend(handles=[line1, line2], loc='best') 
""" Alternativ loc='lower right' """

plt.grid(color='aqua', alpha=0.6, linestyle='dashdot')
plt.show()
```
