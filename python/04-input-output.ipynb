## Dateien mit `open()` √∂ffnen



```python
textfilepath = "data/myTextData.txt"

# Datei sicher √∂ffnen
with open(textfilepath, "r") as my_file:
    # Inhalt der Datei lesen und als Objekt speichern
    my_file_content = my_file.read()
print(my_file_content)
```

```python
with open(textfilepath, "r") as my_file:
    # √ºber jede Zeile der Datei iterieren
    for cur_line in my_file:
        print("This line: {}".format(cur_line))
```

```python
with open(textfilepath, "r") as my_file:
    my_line = my_file.readline()
print(my_line)
```

```python
with open(textfilepath, "r") as my_file:
    my_lines = my_file.readlines()
print(my_lines[3])
```

# Logging

```python
import logging

# logging-Datei spezifizieren
logging.basicConfig(filename="data/my_log.log", filemode="a", format="%(asctime)s - %(process)d-%(levelname)s-%(message)s")

item = 'Kaffeemaschine'

# Logging-Nachrichten mit unterschiedlichen Levels
# der Ereignisschwere erzeugen
logging.debug("Nachricht an mich selbst: Besser bitte")
logging.info("Dies ist ein Test-Skript")
logging.warning("Der Kaffee ist fast leer")
logging.error("Der Kaffee ist leer")
logging.critical('%s hat einen kritischen Fehler verursacht', item)
```
