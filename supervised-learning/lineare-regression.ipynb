# Lineare Regression

## Einfaches Lineares Regressionsmodell

### Daten festlegen


```python
data = {
    'Jahr': [2011, 2012, 2013, 2014, 2015],
    'Einnahmen': [50,   54,   58,   55,   60]
}

predict_data = [2017]
columns      = ['Jahr']
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(data)

plt.scatter(df['Jahr'], df['Einnahmen'])
plt.title('Einnahmen eines fiktiven Unternehmens')
plt.xlabel('Jahr')
plt.ylabel('Einnahmen in Millionen USD')
plt.xticks(df['Jahr'])
plt.grid(True)
```

### Vorhersagemodell erstellen


```python
n = len(df['Jahr'])
df['x2'] = df['Jahr'] ** 2
df['xy'] = df['Jahr'] * df['Einnahmen']
sum_x = df['Jahr'].sum()
sum_y = df['Einnahmen'].sum()
sum_x2 = df['x2'].sum()
sum_xy = df['xy'].sum()

w1 = (n * sum_xy - sum_y * sum_x) / \
    (-(sum_x)**2 + n * sum_x2)

w0 = 1/n * sum_y - w1/n * sum_x

print(f"y = {round(w0, 4)} + {round(w1, 4)}x")
```

### Vorhersage 

```python
new_row = pd.DataFrame(predict_data, columns=columns)
df = pd.concat([df, new_row], ignore_index=True)

df['y_pred'] = w0 + w1 * df['Jahr']

plt.scatter(df['Jahr'], df['Einnahmen'], label='Einnahmen')
plt.plot(df['Jahr'], df['y_pred'], label='Vorhersage')
plt.title('Einnahmen und Einnahmevorhersagen eines fiktiven Unternehmens')
plt.xlabel('Jahr')
plt.ylabel('Millionen USD')
plt.xticks(df['Jahr'])
plt.legend()
plt.grid(True)
```
