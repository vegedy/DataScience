# Kerndichtesch√§tzung

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

x_grid = np.linspace(-5, 5, 500)

def gaussian_kernel(u):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * u**2)

def kde(x_grid, data, h):
    return np.array([
        np.mean(gaussian_kernel((x - data) / h)) / h
        for x in x_grid
    ])

def plot_kde(h):
    density = kde(x_grid, data, h)
    plt.plot(x_grid, density, label=f"KDE (h={h})")

plot_kde(0.05)
plot_kde(0.1)
plot_kde(0.2)

plt.hist(data, bins=30, density=True, alpha=0.3, label='Histogramm')
plt.legend()
plt.title(f"Kerndichtesch√§tzung mit Gau√ü-Kernel")
plt.show()
```
