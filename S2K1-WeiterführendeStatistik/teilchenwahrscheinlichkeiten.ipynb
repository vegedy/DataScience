```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb

# Anzahl der Teilchen und Positionen
n_particles = 10
n_positions = 5

# Wahrscheinlichkeiten fr die Verteilungen
probabilities = []

# Berechnung der Wahrscheinlichkeiten fr jede mgliche Verteilung
for k in range(n_positions + 1):
    if k * 2 <= n_particles:  # Nur gltige Verteilungen
        # Anzahl der Mglichkeiten, k Positionen mit 2 Teilchen zu whlen
        ways = comb(n_positions, k) * (comb(n_particles, 2) ** k) * comb(n_particles - 2 * k, n_particles - 2 * k)
        # Gesamtzahl der mglichen Konfigurationen
        total_configurations = n_positions ** n_particles
        # Wahrscheinlichkeit
        probability = ways / total_configurations
        probabilities.append(probability)
    else:
        probabilities.append(0)

# Visualisierung
x_labels = [f"{i} Positionen mit 2 Teilchen" for i in range(n_positions + 1)]
x = np.arange(len(x_labels))

plt.figure(figsize=(10, 6))
plt.bar(x, probabilities, color='skyblue')
plt.xticks(x, x_labels, rotation=45)
plt.ylabel('Wahrscheinlichkeit')
plt.title('Wahrscheinlichkeiten fr die Verteilung von 10 Teilchen auf 5 Positionen')
plt.ylim(0, max(probabilities) * 1.1)  # Setze die y-Achse etwas hher als die maximale Wahrscheinlichkeit
plt.grid(axis='y')

# Zeige die Werte ber den Balken an
for i, v in enumerate(probabilities):
    plt.text(i, v + 0.01, f"{v:.6f}", ha='center')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb

# Anzahl der Teilchen und Positionen
n_particles = 10
n_positions = 5

# Wahrscheinlichkeiten fr die Verteilungen
probabilities = []

# Berechnung der Wahrscheinlichkeiten fr jede mgliche Verteilung
for k in range(n_positions + 1):
    if k * 2 <= n_particles:  # Nur gltige Verteilungen
        # Anzahl der Mglichkeiten, k Positionen mit 2 Teilchen zu whlen
        ways = comb(n_positions, k) * (comb(n_particles, 2) ** k) * comb(n_particles - 2 * k, n_particles - 2 * k)
        # Gesamtzahl der mglichen Konfigurationen
        total_configurations = n_positions ** n_particles
        # Wahrscheinlichkeit
        probability = ways / total_configurations
        probabilities.append(probability)
    else:
        probabilities.append(0)

# Visualisierung
x_labels = [f"{i} Positionen mit 2 Teilchen" for i in range(n_positions + 1)]
x = np.arange(len(x_labels))

plt.figure(figsize=(10, 6))
plt.bar(x, probabilities, color='skyblue')
plt.xticks(x, x_labels, rotation=45)
plt.ylabel('Wahrscheinlichkeit (logarithmisch)')
plt.title('Wahrscheinlichkeiten fr die Verteilung von 10 Teilchen auf 5 Positionen')
plt.yscale('log')  # Setze die y-Achse auf logarithmisch
plt.grid(axis='y')

# Zeige die Werte ber den Balken an
for i, v in enumerate(probabilities):
    if v > 0:  # Nur Werte grer als 0 anzeigen
        plt.text(i, v * 1.1, f"{v:.6f}", ha='center')

plt.tight_layout()
plt.show()
```
