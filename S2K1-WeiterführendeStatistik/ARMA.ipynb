# ARMA

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter
phi1 = 0.7
theta1 = 0.5
n = 30  # Anzahl der Schritte

# Initialwerte
x = [1]
eps = [1]

# ARMA(1,1) erzeugen
for t in range(1, n):
    eps.append(0)  # danach kein Rauschen mehr
    x_new = phi1 * x[t-1] + theta1 * eps[t-1] + eps[t]
    x.append(x_new)

# Plot
plt.figure(figsize=(10, 4))
plt.plot(x, marker='o')
plt.title(f"ARMA(1,1) mit $\\phi_1={phi1}, \\theta_1={theta1}$")
plt.xlabel('Zeit t')
plt.ylabel('X(t)')
plt.grid(True)
plt.show()
```

```python
# Parameter
phi1, phi2 = 0.5, 0.4
theta1, theta2 = 0.4, 0.2
n = 30

# Initialwerte
x = [1, 1]
eps = [1, 1]

# ARMA(2,2) erzeugen
for t in range(2, n):
    eps.append(0)
    x_new = (phi1 * x[t-1] + phi2 * x[t-2] +
             theta1 * eps[t-1] + theta2 * eps[t-2] +
             eps[t])
    x.append(x_new)

# Plot
plt.figure(figsize=(10, 4))
plt.plot(x, marker='o')
plt.title(f"ARMA(2,2) mit $\\phi_1={phi1}, \\phi_2={phi2}, \\theta_1={theta1}, \\theta_2={theta2}$")
plt.xlabel('Zeit t')
plt.ylabel('X(t)')
plt.grid(True)
plt.show()
```
